Breadth First Search (BFS)

```python
def bfs(graph, source):
  visited = set() #untuk melacak node yang sudah dikunjungi
  bfs_traversal = list() #hasil traversal BFS
  queue = list() #queue

  # push node root ke antrian dan tandai sebagai telah dikunjungi
  queue.append(source)
  visited.add(source)


# loop sampai antrian kosong
  while queue:
    # pop node depan antrian dan tambahkan ke bfs_traversal
    current_node = queue.pop(0)
    bfs_traversal.append(current_node)

    # periksa semua node neighbour dari node saat ini
    for neighbour_node in graph[current_node]:    
        # jika node neighbour belum dikunjungi,
        # Dorong mereka ke antrean dan tandai sebagai telah dikunjung
        if neighbour_node not in visited:
            visited.add(neighbour_node)
            queue.append(neighbour_node)
  return bfs_traversal

def main():
    graph = {
        'A': ['B', 'C'],
        'B': ['A', 'D', 'E'],
        'C': ['A', 'F', 'G'],
        'D': ['B'],
        'E': ['B'],
        'F': ['C'],
        'G': ['C']
    }

    bfs_traversal = bfs(graph, 'A')
    print(f"BFS: {bfs_traversal}")

if __name__=='__main__':
    main()
```

Depth First Search (DFS)

```python
def dfs(graph, source, visited, dfs_traversal):
    if source not in visited:
        dfs_traversal.append(source)
        visited.add(source)

        for neighbor_node in graph[source]:
            dfs(graph, neighbor_node, visited, dfs_traversal)

    return dfs_traversal

def main():
    visited = set()
    dfs_traversal = list()

    graph = {
        'A': ['B', 'C'],
        'B': ['D', 'E'],
        'C': ['F', 'G'],
        'D': [],
        'E': [],
        'F': [],
        'G': []
    }

    print(f"DFS: {dfs(graph, 'A', visited, dfs_traversal)}")

main()
```

Iterative Deepening Depth First Search

```python
# Program Pythom untuk mencetak traversal DFS dari yang diberikan
# diberikan grafik
from collections import defaultdict

# Kelas ini merepresentasikan graf berarah menggunakan adjacency
# daftar representasi
class Graph:

  def __init__(self,vertices):

    # Jumlah simpul(vertices)
    self.V = vertices

    # kamus default untuk menyimpan grafik
    self.graph = defaultdict(list)

  # berfungsi untuk menambahkan tepi(edge) ke grafik
  def addEdge(self,u,v):
    self.graph[u].append(v)

  # Fungsi untuk melakukan pencarian Kedalaman-Terbatas
  # dari sumber yang diberikan 'src'
  def DLS(self,src,target,maxDepth):

    if src == target : return True

  # Jika mencapai kedalaman maksimum, hentikan pengulangan.
    if maxDepth <= 0 : return False

    # Recur untuk semua simpul yang berdekatan dengan simpul ini
    for i in self.graph[src]:
        if(self.DLS(i,target,maxDepth-1)):
          return True
    return False

  # IDDFS untuk mencari jika target dapat dijangkau dari v.
  # Ini menggunakan DLS rekursif ()
  def IDDFS(self,src,target,maxDepth):

    # Berulang kali pencarian batas kedalaman hingga
    # kedalaman maksimum
    for i in range(maxDepth):
      if (self.DLS(src, target, i)):
        return True
    return False

# Buat grafik yang diberikan dalam diagram di atas
g = Graph (7);
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(0, 4)
g.addEdge(1, 3)
g.addEdge(1, 5)
g.addEdge(2, 6)   

target = 6; maxDepth = 3; src = 0

if g.IDDFS(src, target, maxDepth) == True:
  print ("Target is reachable from source " + 
     "within max depth")
else :
  print ("Target is NOT reachable from source " + 
     "within max depth")
```

```python
#Program Python3 untuk BFS Dua Arah
# Cari ke jalur pemeriksaan antara dua simpul

# Definisi kelas simpul fpr ke
# ditambahkan ke grafik
class AdjacentNode:

  def __init__(self, vertex):

    self.vertex = vertex
    self.next = None

# Implementasi BidirectionalSearch
class BidirectionalSearch:

  def __init__(self, vertices):

    # Initialize vertices and
    # graph with vertices
    self.vertices = vertices
    self.graph = [None] * self.vertices

    # Inisialisasi queue untuk maju(forward)
    # tambahkan pencarian mundur(backward)
    self.src_queue = list()
    self.dest_queue = list()

    # Menginisialisasi sumber dan
    # node tujuan yang dikunjungi sebagai Salah
    self.src_visited = [False] * self.vertices
    self.dest_visited = [False] * self.vertices

    # Initializing source and destination
    # parent nodes
    self.src_parent = [None] * self.vertices
    self.dest_visited = [None] * self.vertices

# Fungsi untuk menambahkan tepi yang tidak diarahkan
  def add_edge(self, src, dest):

    # Tambahkan tepi(edge) ke grafik

    # Tambahkan sumber(sourc) ke tujuan
    node = AdjacentNode(dest)
    node.next = self.graph[src]
    self.graph[src] = node

    # Karena grafik tidak diarahkan, tambahkan
    # tujuan ke sumber(sourc)
    node = AdjacentNode(src)
    node.next = self.graph[dest]
    self.graph[dest] = node

  # Fungsi untuk Breadth First Search
  def bfs(self, direction = 'forward'):

    if direction == 'forward':

      # BFS ke arah depan
      current = self.src_queue.pop(0)
      connected_node = self.graph[current]

      while connected_node:
        vertex = connected_node.vertex

        if not self.src_visited[vertex]:
          self.src_queue.append(vertex)
          self.src_visited[vertex] = True
          self.src_parent[vertex] = current

        connected_node = connected_node.next
      else:

        # BFS ke arah belakang
        current = self.dest_queue.pop(0)
        connected_node = self.graph[current]

        while connected_node:
          vertex = connected_node.vertex

          if not self.dest_visited[vertex]:
            self.dest_queue.append(vertex)
            self.src_visited[vertex] = True
            self.src_parent[vertex] = current

          connected_node = connected_node.next

      # Periksa simpul(vertex) yang berpotongan
      def is_intersecting(self):

        # Mengembalikan simpul yang berpotongan
        # jika ada yang lain -1
        for i in range(self.vertices):
          if (self.src_visited[i] and
            self.dest_visited[i]):
            return i

        return -1
      # Cetak jalur(path) dari sumber(source) ke target
      def print_path(self, intersecting_node, src, dest):

        # Cetak jalur(path) terakhir dari
        # sumber ke tujuan
        path = list()
        path.append(intersecting_node)
        i = intersecting_node

        while i != dest:
          path.append(self.dest_parent[i])
          i = self.dest_parent[i]

        print("*****Path*****")
        path = list(map(str, path))

        print(' '.join(path))

      # Function for bidirectional searcing
      def bidirectional_search(self, src, dest):

        # Tambahkan sumber(src) ke antrean(queue) dan tandai
        # dikunjungi sebagai Benar dan tambahkan
        # induk(parent) sebagai -1
        self.src_queue.append(dest)
        self.dest_visited[dest] = True
        self.dest_parent[dest] = -1

        while self.src_queue and self.dest_queue:

          # BFS dalam arah maju dari
          # Sumber Vertex
          self.bfs(direction = 'forward')

          # BFS dalam arah terbalik
          # dari Vertex Tujuan
          self.bfs(direction = 'backward')

          # Check for intersecting vertex
          intersecting_node = self.is_intersecting()

          # Jika simpul(vertex) berpotongan ada
          # jalur(path) dari sumber(sourc) ke
          # tujuan(destination) ada
          if intersecting_node != -1:
            print(f"Path exists between {src} and {dest}")
            print(f"Intersecting at : {intersecting_node}")
            self.print_path(intersecting_node, src, dest)
            exit(0)
          return -1
        
      # Driver code
      # Kode pengemudi
     
    
    # Number of Vertices in graph
    n = 15

    # Source Vertex
    src = 0

    # Destination Vertex
    dest = 6

    # Create a graph
    graph = BidirectionalSearch(n)
    graph.add_edge(0, 4)
    graph.add_edge(1, 4)
    graph.add_edge(2, 5)
    graph.add_edge(3, 5)
    graph.add_edge(4, 6)
    graph.add_edge(5, 6)
    graph.add_edge(6, 7)
    graph.add_edge(7, 8)
    graph.add_edge(8, 9)
    graph.add_edge(8, 10)
    graph.add_edge(9, 11)
    graph.add_edge(9, 12)
    graph.add_edge(10, 13)
    graph.add_edge(10, 14)

    out = graph.bidirectional_search(src, dest)

    if out == -1:
      print(f"Path does not exist between {src} and {dest}")
```
